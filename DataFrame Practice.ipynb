```python
def get_countries_gpd() -> DataFrame:
    df1 = DataFrame(
        index= ['USA', 'Russia', 'China', 'South Korea', 'France', 'Germany', 'UK', 'Japan', 'Canada', 'Antarctica', 'Arctica'],
        data={
            'gdp' : [28781, 2056, 18532, 1760, 3130, 4591, 3495, 4110, 2242, 0, 0],
            'population' : [335.8, 146.1, 1409.6, 51.2, 68.4, 84.673, 67.5, 123.8, 41.0, 0.1, 0.2],
            'region': ['America', 'Europe', 'Asia', 'Asia', 'Europe', 'Europe','Europe', 'Asia', 'America','Antarctica', 'Arctica']
        })
    return df1

df1=get_countries_gpd()
df1

# assert isinstance(df1['gdp'], Series) #Проверяем существование колонки. Колонка - Series!
avg_gdp = df1['gdp'].mean() #Находим средний ВВП
# print(round(avg_gdp, 2))

filtered_df1 = df1[df1['gdp'] <= avg_gdp] #Фильтруем по странам, где ВВП меньше среднего значения
#filtered_df1

real_countries = df1[(df1['gdp'] <= avg_gdp) & ((df1['region'] != 'Arctica') & (df1['region'] != 'Antarctica'))] #Берём только тех у кого ВВП выше среднего и исключаем Арктику с Антарктикой

real_countries_2 = df1[(df1['gdp'] <= avg_gdp) & ~ ((df1['region'] == 'Arctica') | (df1['region'] == 'Antarctica'))] #То же, что и real_countries, но через другие выражения (Булевы)

real_countries
```

```python
# Вычисления как в Excel с помощью Pandas
def get_sales() -> DataFrame:
    return DataFrame(
        data={
            'quantify': [100, 123, 134, 56, 99, 64, 38, 98, 192, 283],
            'price': [12, 14, 18, 17, 19, 21, 20, 23, 19, 17],
            'discount': [0.01, 0.02, 0, 0, 0.03, 0.04, 0.05, 0.01, 0.015, 0.025]
        },
        index=['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct']
    )
sales=get_sales()

sales['profit'] = sales['price']* sales['quantify'] #Добавляем новую колонку и вычисляем значения в ней
sales['price_after_discount'] = sales['price']* (1 - sales['discount']) #Добавляем и вычисляем ещё одну колонку
sales['profit_after_discount'] = sales['price_after_discount']* sales['quantify'] #То же

rub_per_usd = 90
sales['profit_in_rub'] = sales['profit']*rub_per_usd #Выражаем в рублях (умножаем столбец на константу)
# print(sales)
sales['profit_in_rub']
```

```python
from pandas import Series, DataFrame
df3=get_countries_gpd()

def get_gdp_per_capita(country_raw: Series,
                       currency: str | None = None,
                       ex_rate: str | None = None) -> float:
    per_capita = country_raw['gdp'] /country_raw['population']
    if currency == 'USD' or currency is None:
        return per_capita
    else:
        return per_capita * ex_rate #Создаём функцию, которая вычисляет ввп на душу населения и может переводить в разные денежные величины

"""Создаём новую колонку с помощью функции. axis='columns' - значит функция применяется вдоль колонок (т.е. к каждому ряду)"""
df3['gdp_per_capita_EUR'] = df3.apply(get_gdp_per_capita, axis='columns', currency='EUR', ex_rate=0.9)


"""Преимущество применения функции - можно передавать доп. параметры, которые нельзя добавить при простом перемножении/делении и тд колонки на колонку"""
df3['gdp_per_capita_EUR_rounded'] = round(df3['gdp_per_capita_EUR'], 2)
df3
```

```python
# Присоединяем DataFrame-ы друг с другом
import pandas as pd
test_result = DataFrame(data=[
    [1, 'math', 'ivan', 5],
    [2, 'chem', 'peter', 4],
    [3, 'bio', 'nick', 5],
    [4, 'hist', 'kate', 4]
    ],
    columns=['key', 'course_id', 'student_id', 'grade']
)
courses3= DataFrame(data=[
    ['math', 'Mathematics'],
    ['chem', 'Chemistry'],
    ['bio', 'Biology'],
    ['hist', 'History']
], columns=['course_id', 'course_name'])
courses3.set_index('course_id', inplace=True)

students = DataFrame(data=[
    ['ivan', 'Иван'],
    ['peter', 'Пётр'],
    ['nick', 'Николай'],
    ['kate', 'Екатерина']
], columns=['student_id', 'student_name'])
students.set_index('student_id', inplace=True)

test_results_with_students=pd.merge(left=test_result, right=students, left_on='student_id', right_on='student_id', how='left')
"""Если колонки-ключи называются одинаково в обоих ДатаФреймах, то можно просто написать on='student_id'"""
# test_results_with_students
final_results = pd.merge(left=test_results_with_students, right=courses3, left_on='course_id', left_index=False, right_index=True, how='inner')
"""Такой синтаксис уместен, когда одну таблицу мы хотим присоединить просто по столбцу, а вторую - по индексу. Хотя здесь бы работал такой же синтаксис, как строкой выше"""
final_results
```

```python
import pandas as pd
import numpy as np

# Создаем пример данных о продажах
data_0 = {
    'Город': ['Москва', 'Москва', 'Москва', 'СПб', 'СПб', 'СПб'],
    'Месяц': ['Янв', 'Фев', 'Мар', 'Янв', 'Фев', 'Мар'],
    'Продажи': [100, 120, 130, 80, 90, 95],
    'Прибыль': [20, 25, 30, 15, 18, 20]
}
datfr = pd.DataFrame(data_0)
# print("Исходные данные:")
# print(datfr)

# Создаем мультииндекс из столбцов 'Город' и 'Месяц'
df_multi = datfr.set_index(['Город', 'Месяц'])
# print(df_multi)

# unstack() по умолчанию разворачивает последний уровень индекса. Т.е. здесь он развернет горизонтально месяцы
df_unstack=df_multi.unstack()
# print(df_unstack)

# Можно указать, какой уровень индекса разворачивать. Здесь развернёт города
df_unstacked_city = df_multi.unstack(level='Город')
# print(df_unstacked_city)

#stack() преобразует столбцы в уровни индекса.
df_restored=df_unstack.stack()
# print(df_restored)

# Создаем данные с мультииндексом в колонках
columns = pd.MultiIndex.from_tuples([
    ('Продажи', 'Январь'),
    ('Продажи', 'Февраль'),
    ('Прибыль', 'Январь'),
    ('Прибыль', 'Февраль')
])

data = [
    [100, 120, 20, 25],  # Москва
    [80, 90, 15, 18]     # СПб
]

df_wide = pd.DataFrame(data, index=['Москва', 'СПб'], columns=columns)
print("\nШирокий формат данных:")
print(df_wide)

# Применяем stack() - преобразует индекс месяцев в колонку
df_stacked = df_wide.stack()
print("\nПосле stack():")
print(df_stacked)
```
